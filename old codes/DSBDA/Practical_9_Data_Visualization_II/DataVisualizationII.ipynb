```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = sns.load_dataset('titanic')
```

```python
print("-----------Dataset first 5 rows-------------")
print(ds.head())
print("\n")
```

```python
print("-------Boxplot Gender vs Age---------------")
sns.boxplot(x='sex', y='age', data=ds)
plt.show()
print("\n")
```

The first quartile starts at around 5 and ends at 22 which means that 25% of the passengers are aged between 5 and 25.<br>
The second quartile starts at around 23 and ends at around 32 which means that 25% of the passengers are aged between 23 and 32.<br>
Similarly, the third quartile starts and ends between 34 and 42, hence 25% passengers are aged within this range and finally the fourth or last quartile starts at 43 and ends around 65.


```python
print("-------------Survived Passengers---------------------")
sns.boxplot(x='sex', y='age', data=ds, hue='survived')
plt.show()
print("\n")
```
